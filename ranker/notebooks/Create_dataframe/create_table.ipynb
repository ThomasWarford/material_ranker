```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from pandas.plotting import table
```

```python
df = pd.read_csv("df_cut_average_drop=0.6944_MS5_MC6_PERPLXTY40.csv", index_col="material_id")
```

```python
df.columns.to_list()
```

```python
def round_(number):
    return round(number, 5)
df["synth_score_2"] = df.synth_score_2.apply(round_)
df["average_drop"] = df.average_drop.apply(round_)
df["member_strength"] = df.member_strength.apply(round_)

df.head()
```

```python
df_sorted = df.sort_values(by=['labels','member_strength'], ascending=[True, False])

#df_sorted = df_sorted[df_sorted.labels==-1]
#N_clusters = 3

N_clusters = len(np.unique(df_sorted.labels))

fig, axs = plt.subplots(N_clusters, 1, figsize=(30,5*N_clusters))

for i in range(0, N_clusters-1):
    df_part1 = df_sorted[df_sorted.labels==i][["formula", "sublattice_element", "average_drop", "synth_score_2", "member_strength", 
                                               "experimentally_observed", "matched_to_a_sc", "crystal_system", "space_group_symbol", 
                                               "point_group"]]
    # min table size of 15
    # if(len(df_part1)>=15):
    #     df_part1 = df_part1.head(15)
    
    axs[i].axis("off")
    axs[i].axis("off")
    #axs[i].set_ylabel('Cluster_ID={0}'.format(i))
    df_part1_with_index_label = df_part1.rename_axis('Cluster_ID={0}'.format(i)).reset_index()
    #widths = [.05, .02, .04, .04, .04, .03, .03, .05, .03, .03]
    
    table(axs[i], df_part1_with_index_label, loc="center", cellLoc='left', rowLabels="None")  # where df is your data frame
    fig.tight_layout()
#plt.subplots_adjust(wspace=0.2)  
plt.tight_layout()
plt.savefig("final_all_clusters_table_no_unclustered.png")
plt.close()
```

```python
df_sorted = df.sort_values(by=['labels','member_strength'], ascending=[True, False])

N_clusters = len(np.unique(df_sorted.labels))

fig, axs = plt.subplots(1, 1, figsize=(30,220))

df_part1 = df_sorted[df_sorted.labels==-1][["formula", "sublattice_element", "average_drop", "synth_score_2", "member_strength", 
                                           "experimentally_observed", "matched_to_a_sc", "crystal_system", "space_group_symbol", 
                                           "point_group"]]


axs.axis("off")
axs.axis("off")
#axs[i].set_ylabel('Cluster_ID={0}'.format(i))
df_part1_with_index_label = df_part1.rename_axis('Cluster_ID={0}'.format(-1)).reset_index()
#widths = [.05, .02, .04, .04, .04, .03, .03, .05, .03, .03]

table(axs, df_part1_with_index_label, loc="center", cellLoc='left', rowLabels="None")  # where df is your data frame
#fig.tight_layout()
    
#plt.subplots_adjust(wspace=0.2)  
#plt.tight_layout()
plt.savefig("new_big_unclustered_table.png")
plt.close()
```

```python
def simple_formula(string):
    cutoff = 7 # allow AxBy formulae
    return bool(len(string)<cutoff)
```

```python
df_unobserved = df[np.logical_not(df.experimentally_observed)].sort_values(by=["average_drop"])[["formula", "sublattice_element", "average_drop", "synth_score_2", "labels"]]
#df_unobserved["synth_drop_rank"] = (df_unobserved["synth_score_2"] + (1-df_unobserved["average_drop"]) ) / 2
df_unobserved = df_unobserved[df_unobserved.formula.apply(simple_formula)]
df_unobserved = df_unobserved[df_unobserved["synth_score_2"]==1.0]
df_unobserved = df_unobserved.sort_values(by=["average_drop"], ascending=True)
df_unobserved.head(15)
```
